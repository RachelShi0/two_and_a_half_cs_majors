```python
import git
import utilities as ut
import pandas as pd
import numpy as np
from timeit import default_timer as timer

# Loading in data
repo = git.Repo("./", search_parent_directories=True)
homedir = repo.working_dir
data = pd.read_csv(f"{homedir}/data/us/covid/nyt_us_counties.csv")
demog = pd.read_csv(f"{homedir}/data/us/demographics/county_populations.csv")
out_dates = ut.all_output_dates()
out_fips, all_row_starts = ut.all_output_fips('sample_submission.csv')
num_dates, num_fips = len(out_dates), len(out_fips)
out = np.zeros((num_dates * num_fips, 9))

# Adding a date_processed line for later usage
data['date_processed'] = pd.to_datetime(data['date'].values)
data['date_processed'] = (data['date_processed'] - data['date_processed'].min()) / np.timedelta64(1, 'D')

# Adding Demographic data
demog.rename(columns={'FIPS':'fips'}, inplace=True)
data = pd.merge(data, demog, how='left', on=['fips'])
```

```python
start = timer()

#Iterating through FIPs
for fi, fips in enumerate(out_fips):
    print('Processing FIPS', fips)
    
    # Performs predictions and quantiles
    preds = ut.make_erf_quant_predictions(data, fips)
    out[np.arange(fi, out.shape[0], num_fips)] = preds
        
    # Add in the header line
out_lines = [','.join(['id'] + ['%d' % x for x in np.arange(10, 91, 10)]) + '\n']
    # Add in all other lines one at a time
for row_head, row in zip(all_row_starts, out):
    out_lines.append(','.join([row_head] + ['%.2f' % val for val in row]) + '\n')

# Writing our results to a file
with open('SEIRQD_submission1.csv', 'w') as f:
    f.writelines(out_lines)
f.close()
end = timer()
```

```python
print('It takes ' + str(end - start) + ' seconds to run our model')
```
